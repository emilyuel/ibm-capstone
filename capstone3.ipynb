```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/emilysanchez/Desktop/capstone_project/Survey2.csv')
df.head()
```

```python
#first I need to vizualise the platforms worked with 

df['PlatformHaveWorkedWith'] = df['PlatformHaveWorkedWith'].str.split(';')
df_exploded = df.explode('PlatformHaveWorkedWith')
df_exploded = df_exploded.reset_index(drop=True)
df_exploded.head()
```

```python
!pip install wordcloud matplotlib
```

```python
from wordcloud import WordCloud
```

```python
text = ' '.join(df_exploded['PlatformHaveWorkedWith'].astype(str).tolist())


wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color='black',
    max_words=100,
    colormap='Purples'
).generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.show()
```

```python
# now I need top 10 web frameworks

df['WebframeHaveWorkedWith'] = df['WebframeHaveWorkedWith'].str.split(';')
df_exploded1 = df.explode('WebframeHaveWorkedWith')
df_exploded1 = df_exploded1.reset_index(drop=True)
df_exploded1.head()
```

```python
webframehww_df = df_exploded1['WebframeHaveWorkedWith'].value_counts().reset_index()
webframehww_df.columns = ['Webframe', 'Count']
webframehww_df.head(10)
```

```python
webframehww_df = webframehww_df.sort_values('Count', ascending=False).head(10)

sns.barplot(x='Webframe', y='Count', data=webframehww_df, color = 'red')
plt.title('Top 10 Webframes Respondents Have Worked With')
plt.xlabel('Webframe')
plt.ylabel('Number of Responses')
plt.tight_layout()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Need word cloud for all the platforms 

df['PlatformWantToWorkWith'] = df['PlatformWantToWorkWith'].str.split(';')
df_exploded2 = df.explode('PlatformWantToWorkWith')
df_exploded2 = df_exploded2.reset_index(drop=True)
df_exploded2.head()
```

```python
df['WebframeWantToWorkWith'] = df['WebframeWantToWorkWith'].str.split(';')
df_exploded2 = df.explode('WebframeWantToWorkWith')
df_exploded2 = df_exploded2.reset_index(drop=True)
df_exploded2.head()
```

```python
countries = df['Country'].value_counts().reset_index()
countries.columns = ['Country', 'Count']

import plotly.express as px

fig = px.choropleth(
    data_frame=countries,
    locations='Country',
    locationmode='country names',  # or 'ISO-3' if you use ISO codes
    color='Count',
    color_continuous_scale='Blues',  # or 'Viridis', 'Plasma', etc.
    title='Number of Respondents per Country'
)

fig.show()
```

```python
fig.update_layout(
    width=800,   # Adjust width
    height=500,  # Adjust height
    title_text='Number of Respondents per Country',
    title_x=0.5  # Center the title
)
	
```
