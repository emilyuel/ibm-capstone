since Ages is an object (they are ranges in the survey responses), I will change the ages to be the median value of each category. I will then replace the values in the dataset to be these median ages. 
Note: 'Under 18 years old' will be assumed as 18, '65 years or older' will be assumed as 65, 'Prefer not to say' will be assumed as the median of all these ages (33.0 years old)

```python
unique_ages = df['Age'].unique()
print(unique_ages)
```

```python
df['Age'] = df['Age'].replace('Under 18 years old', 18.0)
df['Age'] = df['Age'].replace('18-24 years old', 21.5)
df['Age'] = df['Age'].replace('25-34 years old', 29.5)
df['Age'] = df['Age'].replace('35-44 years old', 39.5)
df['Age'] = df['Age'].replace('45-54 years old', 49.5)
df['Age'] = df['Age'].replace('55-64 years old', 59.5)
df['Age'] = df['Age'].replace('65 years or older', 65.0)
df['Age'] = df['Age'].replace('Prefer not to say', 33.0)
```

```python
df['Age'] = df['Age'].replace(30.0, 29.5)
mean_age = df['Age'].mean()
print(mean_age)
```

```python
median_age = df['Age'].median()
print(median_age)
```

```python
print(df.isnull().sum()) 
```

```python
df['YearsCode'] = df['YearsCode'].replace('Less than 1 year', 1)
df['YearsCode'] = df['YearsCode'].replace('More than 50 years', 51)
df['YearsCodePro'] = df['YearsCodePro'].replace('Less than 1 year', 1)
df['YearsCodePro'] = df['YearsCodePro'].replace('More than 50 years', 51)
```

```python
# Cleaning the NaNs and replacing them with the mean
df['YearsCode'] = pd.to_numeric(df['YearsCode'], errors='coerce')
df['YearsCode'] = df['YearsCode'].fillna(df['YearsCode'].mean())
print(df['YearsCode'].head())
```

```python
# Cleaning the NaNs and replacing them with the mean
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
df['YearsCodePro'] = df['YearsCodePro'].fillna(df['YearsCodePro'].mean())
print(df['YearsCodePro'].head())
```

```python
avg_years_code = df['YearsCode'].mean()
avg_years_codepro = df['YearsCodePro'].mean()

med_years_code = df['YearsCode'].median()
med_years_codepro = df['YearsCodePro'].median()
```

```python
print(f"Mean years coding: {avg_years_code}")
print(f"Mean years coding professionally: {avg_years_codepro}")

print(f"Median years coding: {med_years_code}")
print(f"Median years coding professionally: {med_years_codepro}")
```

```python
mean_comp = df['ConvertedCompYearly'].mean()
median_comp = df['ConvertedCompYearly'].median()

print(f"Average yearly compensation: {mean_comp}")
print(f"Median yearly compensation: {median_comp}")
```
